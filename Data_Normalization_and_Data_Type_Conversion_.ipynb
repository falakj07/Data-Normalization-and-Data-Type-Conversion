```python
## Dataset
import pandas as pd
import numpy as np

data = {
'Product':['Laptop','Mobile','Tablet','Headphones','Smartwatch','Camera'],
'Price' : [55000,20000,30000,2000,8000,45000],
'Units sold' :[25,60,35,80,50,15],
'Discount':[10,15,20,12,18,8]
}

df = pd.DataFrame(data)
print(df)
```

```python
# Min - Max Normaliztaion for price
df["Price_MinMax"] = (df["Price"] - df["Price"].min()) / (df["Price"].max() - df["Price"].min())

print("\nMinMax Normalised Price")
print(df[['Product', 'Price', 'Price_MinMax']])
```

```python
#2. Z-Score Normalization for Price
df["Units_ZScore"] = (df["Units sold"] - df["Units sold"].mean()) / df["Units sold"].std()

print("\nZ-Score Normalized Units Sold")
print(df[['Product', 'Units sold', 'Units_ZScore']])
```

```python
# Normalize Multiple Columns
cols = ['Price','Units sold','Discount']


df_norm = (df[cols] - df[cols].min()) / (df[cols].max() - df[cols].min())
print("\nMin-Max Normalized DataFrame:")
print(df_norm)
```

```python
#
```

```python
# 3 . Decimal scaling
df["Price_Scaled"] = df["Price"] / 100

print('\nDecimal Scaling')
print(df[['Product', 'Price', 'Price_Scaled']])
```

```python
data = {
    'Order ID': [101, 102, 103, 104, 105, 106],
    'Customer_Gender': ['Male', 'Female', 'Male', 'Female', 'Male', 'Female'],
    "Payment_Method": ['Credit Card', 'PayPal', 'Credit Card', 'PayPal', 'Credit Card', 'PayPal'],
    'Product-category': ['Electronics', 'Clothing', 'Electronics', 'Clothing', 'Electronics', 'Clothing'],
    'City':['New York', 'Los Angeles', 'Chicago', 'San Francisco', 'Miami', 'Seattle'],
    'Order_value': [1500, 800, 2000, 1667, 12788, 7756]
}

df1=pd.DataFrame(data)
df1
```

```python
## Method1 : Label Encoding
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()
df1['Gender_Label'] = le.fit_transform(df1['Customer_Gender'])
df1
```

```python
## Method2 : One - Hot encoding
df_encoded = pd.get_dummies(df1, columns=['Payment_Method'])

df_encoded
```
